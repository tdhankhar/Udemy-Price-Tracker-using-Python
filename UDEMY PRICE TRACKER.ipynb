```python
import requests
from bs4 import BeautifulSoup
import smtplib
import time

def check_price(URL,original_price,user):
    headers = {"User-Agent":'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36'}
    page = requests.get(URL,headers=headers)

    soup = BeautifulSoup(page.content,'html.parser')
    
    scrapped_price = soup.find(class_ = "price-text__current")
    scrapped_price = scrapped_price.get_text().strip()
    new_price = toInt(scrapped_price)
    
    if new_price == original_price :
        print("# Change in Price found")
        send_mail(URL,new_price,user)
    else:
        print("# No Fall in price")
    
def send_mail(URL,new_price,user):
    server = smtplib.SMTP('smtp.gmail.com',587)
    server.ehlo()
    server.starttls()
    server.ehlo()
    server.login('tdhankhar@gmail.com','9868531277')   #sender's id pass
    subject = "Udemy Price Tracker Notification"
    body = f"New Price : {new_price}\nCheck out the given link : {URL}"
    msg = f"Subject : {subject}\n\n{body}"
    server.sendmail(
        "tdhankhar@gmail.com",   #sender
        user,  #reciever
        msg
    )
    print("# E-Mail notification has been sent!")
    server.quit()

def toInt(price):
    converted_price = 0
    for i in price:
        if i=='.':
            break
        elif isInt(i):
            converted_price = converted_price*10+int(i)
        else:
            continue
    return converted_price
    
def isInt(ch):
    if ch=='0' or ch=='1' or ch=='2' or ch=='3' or ch=='4' or ch=='5' or ch=='6' or ch=='7' or ch=='8' or ch=='9':
        return True
    else:
        return False
```

```python
def run_app(URL,original_price,user):
    print("----------App is running-----------")
    while True:
        print("# Checking the Price")
        check_price(URL,original_price,user)
        print("# Waiting for 6hrs")
        #time.sleep()

def setup():
    print("**************WELCOME TO UDEMY PRICE TRACKER*****************")
    user = input("Enter your Email address : ")
    URL = input("Enter the URL of the course you are interested in : ")
    headers = {"User-Agent":'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36'}
    page = requests.get(URL,headers=headers)

    soup = BeautifulSoup(page.content,'html.parser')
    
    title = soup.find(attrs={'class': 'clp-lead__title'})
    title=title.get_text().strip()
    print(f"Course Name : {title}")
    
    scrapped_price = soup.find(class_ = "price-text__current")
    scrapped_price = scrapped_price.get_text().strip()
    original_price = toInt(scrapped_price)
    print(f"Current Price : {original_price}\n")
    
    submit = input("Enter 'y' to finish the setup and 'n' to exit : ")
    if submit=='n':
        return
    elif submit=='y':
        print("Setup complete!")
    else:
        print("wrong input")
        return
    
    print("Thanks for completing the setup!!!\nYou will recieve a mail notification whenever there is any fall in the price")
    run_app(URL,original_price,user)
    
    
    
```
